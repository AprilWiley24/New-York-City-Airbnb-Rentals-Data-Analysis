### 04. Data Cleaning and Consistency Checks

```python
# Check for mixed types
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Correct mixed types
df['name'] = df['name'].astype('str')
df['host_name'] = df['host_name'].astype('str')
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Replace missing values for price column with median
df['price'].fillna(df['price'].median(), inplace=True)
```

```python
# Replace missing values for minimum_nights column with median
df['minimum_nights'].fillna(df['minimum_nights'].median(), inplace=True)
```

```python
# Replace missing values for last_review column with most common date
df['last_review'].fillna(df['last_review'].mode()[0], inplace=True)
```

```python
# Replace missing values for reviews_per_month column with mean
df['reviews_per_month'].fillna(df['reviews_per_month'].mean(), inplace=True)
```

```python
# Check for missing values again
df.isnull().sum()
```

```python
# Create subset to check for duplicates
df_dups = df[df.duplicated()]
```

```python
# Check for duplicates
df_dups
```

```python
# Assign string values to id and host_id columns
df['id'] = df['id'].astype('str')
df['host_id'] = df['host_id'].astype('str')
```

```python
# Rename ambiguous columns 
df.rename(columns = {'id' : 'Listing id'}, inplace = True)
df.rename(columns = {'neighbourhood_group' : 'Borough'}, inplace = True)
df.rename(columns = {'neighbourhood' : 'Neighborhood'}, inplace = True)
df.rename(columns = {'minimum_nights' : 'Minimum nights for rent'}, inplace = True)
df.rename(columns = {'calculated_host_listings_count' : 'Number of listings per host'}, inplace = True)
df.rename(columns = {'availability_365' : 'Availability per year'}, inplace = True)
```

```python
# Rename remaining columns for formatting purposes
df.rename(columns = {'host_id' : 'Host id'}, inplace = True)
df.rename(columns = {'latitude' : 'Latitude'}, inplace = True)
df.rename(columns = {'longitude' : 'Longitude'}, inplace = True)
df.rename(columns = {'room_type' : 'Room type'}, inplace = True)
df.rename(columns = {'price' : 'Price'}, inplace = True)
df.rename(columns = {'number_of_reviews' : 'Number of reviews'}, inplace = True)
df.rename(columns = {'last_review' : 'Last review'}, inplace = True)
df.rename(columns = {'reviews_per_month' : 'Reviews per month'}, inplace = True)
```

```python
# Drop columns not needed for analysis
df = df.drop(columns = ['name'])
df = df.drop(columns = ['host_name'])
```
